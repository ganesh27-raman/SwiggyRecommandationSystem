### Missing Value Analysis

```python
#Drop the columns which is not needed
df.drop(columns=['lic_no','address','menu','link'], inplace=True)
```

```python
# Drop rows where 'name' is NaN or empty
df = df[df['name'].notna() & (df['name'].str.strip() != '')]
df = df[df['cuisine'].notna() & (df['cuisine'].str.strip() != '')]
```

```python
df['rating'] = df['rating'].replace('--', np.nan).astype(float)
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python
df['cost'] = df['cost'].replace({'₹': ''}, regex=True).str.strip().astype(float)
```

```python
def parsing_rating_count(val):
    val = val.strip()
    if val == 'Too Few Ratings':
        return np.nan
    if 'K+' in val:
        return int(float(val.replace('K+ ratings', '')) * 1000)
    elif '+' in val:
        return int(val.replace('+ ratings', ''))
    else:
        return np.nan
    
df['rating_count'] = df['rating_count'].apply(parsing_rating_count)

#Filling the nan values with Median
df['rating_count'].fillna(df['rating_count'].median(), inplace=True)
```

```python
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[col] < lower) | (data[col] > upper)]

outliers_cost = detect_outliers_iqr(df, 'cost')
print(f"Outliers in cost: {len(outliers_cost)}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Assuming df_encoded is your DataFrame and it has 'cost', 'rating', and one-hot encoded columns
def cosine_similarity_recommendation(df_encoded, selected_restaurant):
    # Ensure columns for cost, rating, and encoded values are selected correctly
    # Replace 'city_main_encoded' and 'name_encoded' with the actual encoded columns
    numerical_columns = ['cost', 'rating', 'rating_count']  # Adjust this list based on your encoding
    encoded_columns = [col for col in df_encoded.columns if 'encoded' in col]  # Example for selecting encoded features
    
    # Combine numerical and encoded columns
    X = df_encoded[numerical_columns + encoded_columns]

    # Standardize the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Calculate the cosine similarity between all restaurants
    cosine_sim = cosine_similarity(X_scaled)

    # Create DataFrame with restaurant names as both columns and indices
    cosine_sim_df = pd.DataFrame(cosine_sim, columns=df_encoded['name'], index=df_encoded['name'])
    
    # Get the similarity scores for the selected restaurant
    similarity_scores = cosine_sim_df[selected_restaurant]

    # Sort the restaurants by similarity score (descending order)
    recommended_items = similarity_scores.sort_values(ascending=False).head(6)  # Top 5, including the selected restaurant itself

    return recommended_items.index

# Example of calling the function with a restaurant name
df_encoded = pd.read_csv('D:/Data Science Projects/Swiggy Recommendation System/src/data/encoded_data.csv')  # Make sure the DataFrame is loaded correctly
selected_restaurant = 'Restaurant A'  # Example selected restaurant
recommendations = cosine_similarity_recommendation(df_encoded, selected_restaurant)

print("Recommended Restaurants:", recommendations)
```
